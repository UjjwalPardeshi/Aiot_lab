```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 

```

```python
df = pd.read_csv("IGN games from best to worst.csv")
```

missing values

```python
df_cleaned = df.dropna()
    
# Save the cleaned DataFrame back to a CSV file
df_cleaned.to_csv('cleaned_file_games.csv', index=False)
```

remove redundencies 

```python
# Check the number of rows before removing duplicates

df = pd.read_csv('cleaned_file_games.csv')

num_rows_before = df.shape[0]

# Remove duplicate rows
df = df.drop_duplicates()

# Check the number of rows after removing duplicates
num_rows_after = df.shape[0]

# Print summary
print("Number of duplicate rows removed:", num_rows_before - num_rows_after)
print("Number of rows after removing duplicates:", num_rows_after)
```
