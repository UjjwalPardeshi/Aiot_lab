```python
data = pd.read_csv('Final_cleaned_file_games.csv')
Q1 = data['release_year'].quantile(0.25)
Q3 = data['score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Analyzing outliers
outliers = data[(data['score'] < lower_bound) | (data['release_year'] > upper_bound) ]
print(f"Number of potential outliers: {len(outliers)}")
data['score'] = np.clip(data['release_year'], lower_bound, upper_bound)
```

number of potential outliers

```python
# Load the CSV file into a DataFrame
data = pd.read_csv('Final_cleaned_file_games.csv')

# Calculate the first quartile (Q1) and third quartile (Q3) for each column
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# Calculate the interquartile range (IQR) for each column
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers for each column
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify potential outliers for each column
outliers = (data < lower_bound) | (data > upper_bound)

# Print the number of potential outliers for each column
print("Number of potential outliers in each column:")
print(outliers.sum())
```
